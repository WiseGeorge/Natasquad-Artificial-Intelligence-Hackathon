    ##

        - $Recall=TP/((TP+FN))$     						
    ##
    - **F1:** Provides a balance between precision and recall.

        - $F1=2√ó(Precision √óRecall)/((Precision+Recall))$				
    ##
    - **Confusion Matrix:** Summarizes the classification performance of a classifier.
    ##
    - **ROC (Receiver Operating Characteristics) Curve:** Graphical representation of the relationship between a classifier's sensitivity (true positive rate) and its specificity (false positive rate). 
    ##
    - **AUC (Area Under the Curve) Score:** Measure of the degree to which a classifier can distinguish between classes. A higher AUC score indicates a better ability to differentiate between classes.

#### *Convolutional Neural Networks*
A Convolutional Neural Network (ConvNet/CNN) is a type of Deep Learning algorithm that can process an input image and distinguish between different objects by assigning importance to various aspects of the image through learnable weights and biases. Compared to other classification algorithms, ConvNets require less pre-processing and have the ability to learn filters/characteristics. The architecture of a ConvNet is modeled after the connectivity pattern of neurons in the human brain and the organization of the visual cortex, where individual neurons respond to stimuli in a limited region known as the receptive field, which overlap to cover the entire visual area **(Saha, 2018)**.

The Neocognitron, introduced by Kunihiko Fukushima in 1980, was a pioneering architecture and is widely regarded as the forerunner of modern Convolutional Neural Networks (CNNs). This groundbreaking architecture incorporated the principles of feature extraction, pooling layers, and convolution within a neural network, ultimately leading to recognition or classification. **(Fukushima, 1980)**.

In the 1990s, Yann LeCun and his colleagues introduced the first modern Convolutional Neural Network (CNN), drawing inspiration from the neocognitron. In their paper "Gradient-Based Learning Applied to Document Recognition," they demonstrated the effectiveness of a CNN model that combines simple features into increasingly complex features for recognizing handwritten characters. This seminal paper has been cited over 17,588 times and is widely regarded as a landmark in the field of deep learning **(LeCun et al., 1998)**.

##### Balancing The Dataset
By limiting each class to 500 instances, a better **class balance** will be created, which can help improve the performance of the model on all classes.

```python
def balance_df(df:pd.DataFrame, label_col:str, max_instances:int=500) -> pd.DataFrame:
    """Receive the dataframe of the dataset, the label column and the number of instances, 
    and return a balanced dataframe with the number of instances as the maximum value."""
    balanced_df = pd.DataFrame()
    for label in df[label_col].unique():
        label_df = df[df[label_col] == label]
        if len(label_df) > max_instances:
            label_df = label_df.sample(max_instances)
        balanced_df = pd.concat([balanced_df, label_df])
    return balanced_df
```

```python
def Save_Split_Data(x_train:np.ndarray, y_train:np.ndarray, x_val:np.ndarray, y_val:np.ndarray, x_test:np.ndarray, y_test:np.ndarray, path:str='../Models/Splited Data/') -> str:
    """Save the split sets as .npy files in the directory"""
    file_names = ['x_train.npy', 'y_train.npy', 'x_val.npy', 'y_val.npy', 'x_test.npy', 'y_test.npy']
    split_sets = [x_train, y_train, x_val, y_val, x_test, y_test]

    for item in file_names:
        if os.path.exists(path+item):
            return 'Splited Data Already Exists'

    for i in range(len(split_sets)):
        np.save(path + file_names[i], split_sets[i])
        
    return 'Splited Data Saved Succesfully'
```

#### ***Transfer Learning Trough Xception Architecture***

- The Xception architecture is initialized adding new configurations, input layer 100x100x3, weights were frozen and used the `imagenet` weights which are pretrained on ImageNet database organized according to WordNet hierarchy **(Fei-Fei et al., 2009)**. The output flow is connected to the output of the Xception engine adding a `2D Global Average Pooling` layer. The parameter compression ratio is exponentially high in this type of layer resulting in a 2D dimensionality of the form `(batch_dimension, n_channels)`, which is different to the `Flatten` layer commonly used to feed fully connected layers, which only restructures the matrix to a single dimension **(Keras, 2023)**. Followed by a `Dropout` layer with 80% probability to avoid possible overfitting mainly due to the extension of the pretrained model, and a `Batch Normalization` layer to normalize the inputs. This layer applies transformations that keep the mean output close to 0 and the standard deviation output close to 1 **(Hinton et al., 2012; Ioffe & Szegedy, 2015)**. It is concluded with a `Dense` output layer with 18 neurons referring to the classes in the domain, using the `softmax activation function`.

- The model is compiled using the `Adaptive Moment Estimation (Adam)` optimizer with a `learning rate` of `3e-4`, the loss function `categorical crossentropy`, the metric `accuracy` and a `batch size` of `32`. The values corresponding to `learning rate` and `dropout` were obtained from the fine tune process using `keras tuner`, the hyperparameters search was run on `[0.03, 0.003 0.0003]` and `[0.75, 0.8]` for `learning rate` and `dropout` respectively **(O'Malley et al., 2019)**. The recommended value for the `batch size` parameter in low-performance computing was selected. The learning rate values were chosen because, although the default value for the `Adam` optimizer is `0.001`, a widely used value is `0.0003` **(Sordello et al., 2019)**. Therefore, an interval from 0.03 to 0.0003 was used, decreasing by a factor of 10. High `dropout` values of `0.75` and `0.8` were selected due the amplitud of the pretreined model could lead to overfitting the predictive model.

##### TL Model Building

```python
def TL_Model(hp):

    engine = tf.keras.applications.Xception(
        # Freezing the weights
        include_top=False,
        # Use Imagenet weights
        weights='imagenet',
        # Input Shape
        input_shape=(img_height, img_width, 3)
    )

    x = tf.keras.layers.GlobalAveragePooling2D(name='avg_pool')(engine.output)
    x = tf.keras.layers.Dropout(hp.Choice('dropout', values=[0.75, 0.8]))(x)
    x = tf.keras.layers.BatchNormalization()(x)
    out = tf.keras.layers.Dense(18, activation='softmax', name='dense_output')(x)

    # Build the Keras model
    model = tf.keras.models.Model(inputs=engine.input, outputs=out)

    # Compile the model
    model.compile(
        # Set optimizer to Adam
        optimizer=tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[0.03, 0.003, 0.0003])),
        # Set loss 
        loss='categorical_crossentropy',
        # Set metrics to accuracy
        metrics=['accuracy']
    )
    return model
```

##### Fine Tuning with Keras Tuner

```python
tuner = RandomSearch(
    TL_Model,
    objective='val_accuracy',
    max_trials=5,
    executions_per_trial=3,
    directory='../Models/Keras_Tuner',
    project_name='Fruits_ConvNet Fine_Tuning')

tuner.search_space_summary()

tuner.search(x_train, y_train,
             epochs=2,
             validation_data=(x_val, y_val))

tuner.results_summary()
```

##### Callbacks
Four callbacks will be used during the training process, early stop, learning rate reduction, tensorboard and checkpoint **(Keras, 2023)**:

- **Early Stop:** Stops the training process after no better values are achieved in the loss metric.
- **Learning Rate Reduction:** Reduces the learning rate of the model, set at `3e-4` by monitoring the loss on validation data allowing for better optimization and accuracy of the model on validation data.
- **Tensorboard:** Tool included in the Tensorflow Framework for real-time visualization of all variables and the model's behaviors, favoring the optimization of hyperparameters.
- **Checkpoint:** Allows to save the weights and biases of the model in a certain state as it is trained, in this case it saves each time the model improves depending on the metric `val_accuracy`.

#### ***Deep Convolutional Neural Network From Scratch Inspired in Xception Architecture***
The Deep Convolutional Neural Network is inspired by the Xception Architecture and consists of three main components: an entry flow, a middle flow, and an exit flow. In the entry flow, the input data is processed through several layers of `convolution`, `max pooling`, and `batch normalization` to extract features from the data. A `residual` connection is also introduced at this stage to improve the flow of information through the network.

The middle flow, which is repeated either once or twice depending on a hyperparameter choice, consists of `separable convolution`, `max pooling`, and `batch normalization` layers that further refine the extracted features. The output flow applies another set of `separable convolution`, `max pooling`, and `batch normalization` layers, followed by a `global average pooling` layer to reduce the dimensionality of the data. A `dense` layer with a hyperparameter choice for the number of units is then applied, followed by a `dropout` layer with a hyperparameter choice for the dropout rate to prevent overfitting. Another `batch normalization` layer is applied before the final `dense` layer with a `softmax activation function` to generate the output probabilities for each class. In each `convolutional` layer, the `use_bias` parameter is set to false since the `batch normalization` layer is applied at the end of each convolutional block.

The model is compile using the `Adam` optimizer with a hyperparameter choice for the `learning rate`. The loss function used is `categorical crossentropy`, and `accuracy` is used as the evaluation metric. The following image is a graphical representation of the XICNN_Model. Image Source: The Author.

<p align="center"><img src="../Images/XICNN.png" alt="Model's Workflow" style="width: 600px"/></p>

##### XICNN Model Building

```python
def Xception_Inspire_CNN(hp,input_shape=(img_height,img_width,3), num_classes=18):
    # Entry flow
    inputs = keras.Input(shape=input_shape)
    x = Conv2D(32, 1, strides=2, padding='same', use_bias=False,activation='relu')(inputs)
    x = MaxPooling2D(3, strides=2, padding='same')(x)
    x = BatchNormalization()(x)
    x = Conv2D(64, 1, padding='same', use_bias=False,activation='relu')(x)
    x = MaxPooling2D(3, strides=2, padding='same')(x)
    x = BatchNormalization()(x)

    residual = Conv2D(128, 1, strides=2, padding='same', use_bias=False,activation='relu')(x)
    residual = BatchNormalization()(residual)

    x = SeparableConv2D(128, 3, padding='same', use_bias=False,activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D(3, strides=2, padding='same')(x)
    
    x = Dropout(0.2)(x)
    x = Add()([x, residual])

    # Middle flow (repeated 1 - 2 times instead of 8)
    for i in range(hp.Choice('midflow', values=[1,2])):
        residual = x
        x = SeparableConv2D(256, 3, padding='same', use_bias=False,activation='relu')(x)
        x = MaxPooling2D(3, strides=2, padding='same')(x)
        x = BatchNormalization()(x)
        

    # Exit flow
    residual = Conv2D(512, 1, strides=2, padding='same', use_bias=False,activation='relu')(x)
    residual = BatchNormalization()(residual)

    x = SeparableConv2D(512, 3, padding='same', use_bias=False,activation='relu')(x)
    x = BatchNormalization()(x)
    x = MaxPooling2D(3, strides=2, padding='same')(x)
    x = Add()([x, residual])

    x = GlobalAveragePooling2D()(x)
    x = Dense(hp.Choice('dense_unit', values=[64,128,256]))(x)
    x = Dropout(hp.Choice('dropout', values=[0.3, 0.4, 0.5]))(x)
    x = BatchNormalization()(x)
    outputs = Dense(num_classes, activation='softmax')(x)

    model = keras.Model(inputs=inputs, outputs=outputs)

    # Compile the model
    model.compile(
        # Set optimizer to Adam(0.0003)
        optimizer = tf.keras.optimizers.Adam(learning_rate=hp.Choice('learning_rate', values=[0.003, 0.0003])),
        # Set loss 
        loss = 'categorical_crossentropy',
        # Set metrics to accuracy
        metrics = ['accuracy']
        )

    return model
```

```python
def history_plot(history:pd.DataFrame) -> None:
    """Receive a history and plot loss and validation loss, accuracy and validation accuracy, and learning rate of the history"""
    epochs = len(history)+1
    fig, (ax1,axl,ax2) = plt.subplots(3,1 , figsize=(9,7))

    # Find the best epoch based on minimum validation loss
    best_epoch = np.argmax(history['val_accuracy']) + 1

    # Max Loss Value
    max_loss = max(history['loss'].max(), history['val_loss'].max())
    tick_interval = round(max_loss/8,1)

    ## Loss & Validation Loss Plot
    ax1.plot(history['loss'], color='royalblue', label="Training loss")
    ax1.plot(history['val_loss'], color='indianred', label="Validation loss")
    ax1.set_xticks(np.arange(1, epochs, 1))
    ax1.set_yticks(np.arange(0, max_loss+tick_interval, tick_interval))
    ax1.set_title('Loss and Validation Loss')
    ax1.set_xlabel('')
    ax1.set_ylabel('Loss Value')
    #Best Epoch
    ax1.axvline(best_epoch, linestyle='--', color='red', label='Best Epoch')
    # Legend
    ax1.legend(loc='best', shadow=True)
    
    ## Learning Rate Plot
    axl.plot(history['lr'], color = 'teal', label = 'Learning Rate')
    axl.set_xticks(np.arange(1, epochs, 1))
    axl.set_title('Learning Rate Reduction')
    axl.set_xlabel('')
    axl.set_ylabel('Learning Rate Value')
    axl.legend(loc='best', shadow=True)

    ## Accuracy & Validation Accuracy Plot
    ax2.plot(history['accuracy'], color='royalblue', label="Training accuracy")
    ax2.plot(history['val_accuracy'], color='indianred',label="Validation accuracy")
    ax2.set_xticks(np.arange(1, epochs, 1))
    ax2.set_title('Accuracy and Validation Accuracy')
    ax2.set_xlabel('Epochs')
    ax2.set_ylabel('Accuracy Value')
    #Best Epoch
    ax2.axvline(best_epoch, linestyle='--', color='red', label='Best Epoch')
    # Legend
    ax2.legend(loc='best', shadow=True)
    
    plt.tight_layout()
    plt.show()
```

As detailed in the model history plots above, both models exhibit robust generalization capabilities, as evidenced by their consistent performance across the training and validation datasets. The benefits of transfer learning: a **higher start**, a **higher slope**, and a **higher asymptote** are evident in the TL_Model history plot **(Torrey & Shavlik, 2009)**. At the end of training, the significant contribution of the callback learning rate reduction to the performance of both models on validation data can be seen. It also shows how the reduction of the learning rate stabilizes the models, keeping them close to the best metrics, the decay of the learning rate is also translated into a smaller amplitude between the training and validation curve. This is due to the fact that the lower the value of the learning rate used in the gradient descent implemented by the Adam optimizer, the smaller its displacement in the search for the global minimum, therefore it can locate this or very close values with greater effectiveness, having less possibility of deviating to less favorable situations **(Geron, 2019)**.
```python
def plot_roc_curves(y_true, TL_yscore, CNN_yscore, n_classes=18):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))
    
    # OvR for TL model
    y_true_bin = label_binarize(y_true, classes=range(n_classes))

    tl_fpr = dict()
    tl_tpr = dict()
    tl_roc_auc = dict()

    for j in range(n_classes):
        tl_fpr[j], tl_tpr[j], _ = roc_curve(y_true_bin[:, j], TL_yscore[:, j])
        tl_roc_auc[j] = auc(tl_fpr[j], tl_tpr[j])

    # Plot ROC curves for each class for TL model
    for j in range(n_classes):
        axes[0].plot(tl_fpr[j], tl_tpr[j], label=f'{CLASS_NAMES[j]} (AUC = {tl_roc_auc[j]:.4f})')

    axes[0].set_xlabel('False Positive Rate')
    axes[0].set_ylabel('True Positive Rate')
    axes[0].set_title('OvR ROC Curve for TL Model')
    axes[0].legend(loc="lower right")
    axes[0].set_ylim(0.7, 1.01)
    
    # OvR for CNN model
    cnn_fpr = dict()
    cnn_tpr = dict()
    cnn_roc_auc = dict()

    for j in range(n_classes):
        cnn_fpr[j], cnn_tpr[j], _ = roc_curve(y_true_bin[:, j], CNN_yscore[:, j])
        cnn_roc_auc[j] = auc(cnn_fpr[j], cnn_tpr[j])

    # Plot ROC curves for each class for CNN model
    for j in range(n_classes):
        axes[1].plot(cnn_fpr[j], cnn_tpr[j], label=f'{CLASS_NAMES[j]} (AUC = {cnn_roc_auc[j]:.4f})')

    axes[1].set_xlabel('False Positive Rate')
    axes[1].set_ylabel('True Positive Rate')
    axes[1].set_title('OvR ROC Curve for CNN Model')
    axes[1].legend(loc="lower right")
    axes[1].set_ylim(0.7, 1.01)

    plt.show()
```
